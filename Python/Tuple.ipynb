# Tuples

```python
# Create your first tuple

tuple1 = ("disco",10,1.2 )
print(tuple1)
```

```python
# Print the type of the tuple created

type(tuple1)
```

```python
# Print the variable on each index

print(tuple1[0])
print(tuple1[1])
print(tuple1[2])
```

```python
# Print the type of value on each index

print(type(tuple1[0]))
print(type(tuple1[1]))
print(type(tuple1[2]))
```

```python
# Use negative index to get the value of the last element

print(tuple1[-1])
print(tuple1[-2])
print(tuple1[-3])
```

# Concatenate Tuples

```python
# Concatenate two tuples

tuple2 = tuple1 + ("hard rock", 10)
print(tuple2)
```

```python
tuple3 = tuple2 + ("the rock" , "jhon cena" , 2 )
print(tuple3)
```

# Slicing

```python
# Slice from index 0 to index 2

tuple2[0:3]
```

```python
# Slice from index 3 to index 4

tuple2[3:5]
```

# length

```python
# Get the length of tuple

len(tuple2)
```

# Nested Tuple

```python
# Create a nest tuple

NestedT =(1, 2, ("pop", "rock") ,(3,4),("disco",(1,2)))

NestedT
```

```python
# Print element on each index, including nest indexes

print("Element 2, 0 of Tuple: ",   NestedT[2][0])
print("Element 2, 1 of Tuple: ",   NestedT[2][1])
print("Element 3, 0 of Tuple: ",   NestedT[3][0])
print("Element 3, 1 of Tuple: ",   NestedT[3][1])
print("Element 4, 0 of Tuple: ",   NestedT[4][0])
print("Element 4, 1 of Tuple: ",   NestedT[4][1])
```

```python
# Print the first element in the second nested tuples

NestedT[2][1][0]
```

```python
# Print the second element in the second nested tuples

NestedT[2][1][1]
```

```python
# Print the first element in the second nested tuples

NestedT[4][1][0]
```

# Immutable Tuple

```python
#Example that tuple are immutable
#immutable means =  unable to be changed.


Demo = ("atharv" , " Rohan" , "kunal" , 12 , 11 , 0)
print(Demo) 
```

```python
#Because it immutable it does not change
Demo[3]="rohit"
```

```python
#but we will change the list in the tuple
```

```python
Demo11 = ("atharv" , " Rohan" , "kunal" , 12 , 11 ,[6, 5],["KO"])
print(Demo11)
```

```python
Demo11[5][0] = 4
print(Demo11)
```

```python
Demo11[6][0] = "OK"
print(Demo11)
```
