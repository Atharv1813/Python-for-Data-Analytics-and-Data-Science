# File Handling

```python
# "r" - Read - Default value. Opens a file for reading, error if the file does not exist

# "a" - Append - Opens a file for appending, creates the file if it does not exist

# "w" - Write - Opens a file for writing, creates the file if it does not exist

# "x" - Create - Creates the specified file, returns an error if the file exists
```

# Appending the file

```python
# Append in the file  "a"

file1 = open("D:\VAR.txt", "a")
file1.write("\n Use the appending.")
file1.close()

file1 = open("D:\VAR.txt", "r")
print(file1.read())
file1.close()
```

# Writing the file

```python
# Write in the file   == "w"

file1 = open("D:\VAR.txt", "w")
file1.write("We using the write.")
file1.close()


file1 = open("D:\VAR.txt", "r")
print(file1.read())
file1.close()
```
