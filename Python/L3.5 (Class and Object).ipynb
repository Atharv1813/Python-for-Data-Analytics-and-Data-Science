# Classes and Objects in Python

```python
# Create a class

class F1Class:
  x = 5
```

```python
# Create an object 

K1 = F1Class()
print(K1.x)
```

```python
# __init__() function built-in function.Use to initialize the attributes of the class

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("John", 36)

print(p1.name)
print(p1.age)
```

```python
class laptop:
    def __init__(self,raml,cpul):
        self.ram = raml
        self.cpu = cpul
        
    def configl(self):
        print("Specification is : ",self.ram," & " , self.cpu)
        
r = laptop(8,"i5")
r.configl()
print(r.ram)
print(r.cpu)
```

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def priname(self):
    print("Hello my name is " + self.name)

p1 = Person("John", 36)
p1.priname()
print(p1.name)
print(p1.age)
```

# Some Methods of Object and class

```python
# The self parameter is a reference to the current instance of the class

class Person:
  def __init__(Place, name, age):
    Place.name = name
    Place.age = age

  def myfunc(Holder):
    print("Hello my name is " + Holder.name)
    print("Hello my age is", Holder.age)

p1 = Person("John", 36)
p1.myfunc()
```

```python
# modify properties on objects

p1.age = 18
p1.myfunc()
```

```python
# Delete the property objects

del p1.age
p1.myfunc()
```

```python
#Delte the object

del p1
p1.myfunc()
```

```python
p1 = Person("The Rock", 37)
p1.myfunc()
```
