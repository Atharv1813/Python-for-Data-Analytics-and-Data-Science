# Copy and Clone List

```python
# Copy (copy by reference) the list A

A = ["hard rock", 10, 1.2]
B = A
print('A:', A)
print('B:', B)
```

```python
# Examine the copy by reference

print('B[0]:', B[0])
A[0] = "banana"
print('B[0]:', B[0])
```

```python
# Clone (clone by value)

demo1 = ["john" , "rock" , "56"]
print(demo1)
```

```python
demo2 = demo1[:]
print(demo2)
```

```python
#clone vlaue will not change

print("demo2[2]:", demo2[2])
demo1[2] = 90
print("demo2[2]:" , demo2[2])
```

# List are Mutable

```python
# Change the element based on the index

C = ["disco", 10, 1.2]
print('Before change:', C)
C[0] = 'hard rock'
print('After change:', C)
```
