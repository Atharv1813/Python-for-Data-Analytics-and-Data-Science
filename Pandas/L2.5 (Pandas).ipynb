# Melt in Pandas

```python
import pandas as pd
df = pd.read_csv("E:\Python\ok\Book4.csv")
df
```

```python
# melt will specifies that value on id_vars denoted
df1 = pd.melt(df, id_vars=["day"])
df1
```

```python
# vae_name change the name of variable and value_name change the name of value.
melted = pd.melt(df, id_vars=["day"], var_name='city', value_name='temperature')
melted
```
