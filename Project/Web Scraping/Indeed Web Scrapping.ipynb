```python
import csv
import requests 
from bs4 import BeautifulSoup
from datetime import datetime



def get_url(position , location ,page):
    position = position.replace(" ","+")
    location = location.replace(" ","+")
    demo1 = "https://in.indeed.com/jobs?q={}&l={}&start={}"
    url = demo1.format(position, location ,page)
    return url




def get_output(in_put):
    atag = in_put.h2.a
    try:
        job_title = atag.get('title')
    except AttributeError:
        job_title = " "
    try:
        location =in_put.find('span', 'location').text.strip()
    except AttributeError:
        location=' '
    try:
        company = in_put.find('span', 'company').text.strip()
    except AttributeError:
        company = '' 
    try:
        job_summary = in_put.find('div', 'summary').text.strip()
    except AttributeError:
        job_summary = ''
    try:
        post_date = in_put.find('span', 'date').text.strip()
    except AttributeError:
        post_date = ''
        
    extract_date = datetime.today().strftime('%Y-%m-%d')
    job_url = 'https://www.indeed.com' + atag.get('href')
    
    
    return (job_title,company ,location, job_summary,  post_date, extract_date, job_url)

    





def main(position , location):
    Output = []
    for l in range(0,121,10):
        url = get_url(position,location,l)
        print(url)
        response = requests.get(url)
        soup = BeautifulSoup(response.text , "html.parser")
        in_puts = soup.find_all("div" , "jobsearch-SerpJobCard")
        for in_put in in_puts:
            Output.append(get_output(in_put))
            
            
    with open('indeed web scrapping.csv','w',newline='',encoding='utf-8') as file:
        a = csv.writer(file)
        a.writerow(['job_title','company','location','job_summary','post_date','extract_date','job_url'])
        a.writerows(Output)

        
```

```python
main("data analyst" , "New Delhi, Delhi")
```
