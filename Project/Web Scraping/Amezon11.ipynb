```python
import csv
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
def get_url(Page):
    url = "https://www.amazon.in/s?k=ultrawide+monitors&page={}".format(Page)
    return url 
```

```python
def get_output(in_put):
    atag = in_put.h2.a
    
    
    try:
        Title = atag.text.strip()
    except AttributeError:
        Title = " "
        
        
    try:
        Ratings = in_put.i.text.strip()
    except AttributeError:
        Ratings=" "
        
        
    try:
        Price = in_put.find('span' , 'a-price-whole').text.strip()
    except AttributeError:
        Price=" "
        
        
    try:
        Priceorg ='https://www.amazon.com'+atag.get('href')
    except AttributeError:
        Priceorg=" "
        
        
        
    return (Title,Ratings, Price,  Priceorg)
        
        
```

```python
def main():
    
    Output = []
    driver = webdriver.Chrome("C:\\Users\\user\\Desktop\\Data Anlytics\\Web Scrapping\\Chrome\\chromedriver")
    for l in range(0,21,1):
        url = get_url(l)
        print(url)
        driver.get(url)
        soup = BeautifulSoup(driver.page_source,'html.parser')
        in_puts = soup.find_all("div" , "sg-col sg-col-4-of-12 sg-col-8-of-16 sg-col-12-of-20")
        for in_put in in_puts:
            Output.append(get_output(in_put))
            
            
                        
            
    with open('Amazon.csv','w',newline='',encoding='utf-8') as file:
        a = csv.writer(file)
        a.writerow(['Title','Ratings','Price','Priceorg'])
        a.writerows(Output)
```

```python
main()
```
